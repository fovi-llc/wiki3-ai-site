# Using the `WebLLM Chat` kernel to run built-in-ai LLMs as JupyterLite kernels

*If you are in the wiki view (static HTML rendering), click "edit" in the top right to open this [JupyterLite](https://jupyterlite.readthedocs.io/) notebook in your browser.*

The `WebLLM Chat` kernel (be sure it is selected in the kernel picker above) is derived from a JupyterLite kernel implemented by [Aleksandar Sekulovski](https://www.linkedin.com/in/aleksandar-s-486829128/) at https://github.com/aleks-s10/agent-client-kernel/tree/main/lite-kernel.  This version currently lives at https://github.com/wiki3-ai/webllm-chat-kernel with Python installer by [Jim White](https://www.linkedin.com/in/jamespaulwhite/) and Copilot.

It is a TS/JS kernel that wraps WebLLM (https://webllm.mlc.ai/,  NPM `@mlc-ai/web-llm`) to run MLC LLMs hosted on HuggingFace ([HuggingFace MLC-LLM models](https://huggingface.co/models?library=mlc-llm))  directly in the browser using [WebGPU](https://en.wikipedia.org/wiki/WebGPU) by [Jakob Hoeg MÃ¸rk](https://www.linkedin.com/in/jakob-hoeg-moerk/).

There is no UX for things like errors/config issues, status, progress, etc yet.  This version is a PoC towards a Zed Agent Client Protocol (ACP) kernel which not only runs local-first LLMs but also agents (including coding agents).

Example using the default, a "small" LLM: [SmolLM2-360M-Instruct-q4f16_1-MLC](https://huggingface.co/mlc-ai/SmolLM2-360M-Instruct-q4f16_1-MLC) which is 360M parameters quantized to 4bit and is 130MB.
