## Using the `WebLLM Chat` kernel to run built-in-ai LLMs as JupyterLite kernels

The `WebLLM Chat` kernel (be sure it is selected in the kernel picker above) is derived from a JupyterLite kernel implemented by [Aleksandar Sekulovski](https://www.linkedin.com/in/aleksandar-s-486829128/) at https://github.com/aleks-s10/agent-client-kernel/tree/main/lite-kernel.  This version currently lives at https://github.com/fovi-llc/built-in-ai-webllm-kernel with Python installer by [Jim White](https://www.linkedin.com/in/jamespaulwhite/) and Copilot.  It will be moving to its own repo RSN.

It is a TS/JS kernel that wraps `@built-in-ai/web-llm` from https://github.com/jakobhoeg/built-in-ai which is an [AI SDK model provider](https://ai-sdk.dev/) for open-source models (using WebLLM) running directly in the browser by [Jakob Hoeg MÃ¸rk](https://www.linkedin.com/in/jakob-hoeg-moerk/).

### &#128679; Under Construction &#128679; ###
There is no UX for things like errors/config issues, status, progress, etc yet.  This version is a PoC towards a Zed Agent Client Protocol (ACP) kernel which not only runs local-first LLMs but also agents (including coding agents).

Example using the default, a "small" LLM: [SmolLM2-360M-Instruct-q4f16_1-MLC](https://huggingface.co/mlc-ai/SmolLM2-360M-Instruct-q4f16_1-MLC) which is 360M parameters quantized to 4bit and is 130MB.

```markdown
how sweet would a rose smell by another name?
```

Switched to [gemma-2-2b-it-q4f16_1-MLC](https://huggingface.co/mlc-ai/gemma-2-2b-it-q4f16_1-MLC).  This takes a while to download and run.  Watch the developer console and/or network tabs to track progress.
